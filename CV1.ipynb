```python
import cv2
import matplotlib.pyplot as plt

```

```python
img = cv2.imread('unnamed.png')
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
print("RGB Image Shape:", img_rgb.shape)
print("grayscale Image Shape:", img_gray.shape)
```

```python
# Display RGB image
plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("RGB Image")
```

```python
# Display Grayscale image
plt.subplot(1, 2, 2)
plt.imshow(img_gray, cmap='gray')
plt.title("Grayscale Image")

plt.show()
```

```python
blur = cv2.GaussianBlur(img_rgb, (5, 5), 0)
```

```python
import numpy as np
```

```python
kernel_sharpen = np.array([[0, -1, 0],
                           [-1, 5,-1],
                           [0, -1, 0]])
sharpen = cv2.filter2D(img_rgb, -1, kernel_sharpen)
```

```python
edges = cv2.Canny(img_rgb, 100, 200)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1, 4, 1); plt.imshow(img_rgb); plt.title("Original")
plt.subplot(1, 4, 2); plt.imshow(blur); plt.title("Blur")
plt.subplot(1, 4, 3); plt.imshow(sharpen); plt.title("Sharpen")
plt.subplot(1, 4, 4); plt.imshow(edges, cmap='gray'); plt.title("Edge Detection")
plt.tight_layout(); plt.show()
```

```python
flip = cv2.flip(img_rgb, 5)
print("Flipped Image Shape:", flip.shape)
plt.imshow(flip)
plt.title("Flipped Image")
```

```python
flip_v = cv2.flip(img_rgb, 0)
flip_h = cv2.flip(img_rgb, 1)
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1); plt.imshow(flip_v); plt.title("Vertical Flip")
plt.subplot(1, 3, 2); plt.imshow(flip_h); plt.title("Horizontal Flip")
```

```python
padded = cv2.copyMakeBorder(img_rgb, 10, 10, 20, 20, cv2.BORDER_CONSTANT, value=[255, 255, 255])
plt.subplot(1, 3, 3); plt.imshow(padded); plt.title("Padded Image")
plt.tight_layout()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a blank image
img = np.zeros((500, 500, 3), dtype=np.uint8)
img[:] = (255, 255, 255)  # Make it white

# Draw a blue line
cv2.line(img, (50, 50), (450, 50), (255, 0, 0), 3)

# Draw a green rectangle
cv2.rectangle(img, (50, 100), (200, 200), (0, 255, 0), 2)

# Draw a red circle
cv2.circle(img, (300, 150), 50, (0, 0, 255), -1)  # -1 = filled

# Draw an ellipse
cv2.ellipse(img, (300, 350), (80, 40), 0, 0, 180, (255, 0, 255), 2)

# Draw text
cv2.putText(img, 'Computer Vision', (100, 450), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2)

# Show image with matplotlib
plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Shapes and Text on Image")
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image and convert to grayscale
img = cv2.imread('profile.jpeg')  # Use your image path here
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply thresholding
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Find contours
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on a copy of the original image
img_copy = img.copy()
cv2.drawContours(img_copy, contours, -1, (0, 255, 0), 2)

# Display result
plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB))
plt.title("Contours Detected")
plt.axis("off")
plt.show()
```
